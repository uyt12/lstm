```python
#splitting data into Training and Testing
data_training = pd.DataFrame(df['Close'][0:int(len(df)*0.70)])
data_testing = pd.DataFrame(df['Close'][int(len(df)*0.70):int(len(df))])
print(data_training.shape)
print(data_testing.shape)
```

```python
data_training.head()
```

```python
data_testing.head()
```

```python
#scaling down the data b/w 0 to 1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
#scaling down the training data
data_training_array = scaler.fit_transform(data_training)
data_training_array
```

```python
#features: 100 days (steps)
x_train=[]
#predictors (the 101th day)
y_train=[]

for i in range(100,data_training_array.shape[0]):
    x_train.append(data_training_array[i-100:i])
    y_train.append(data_training_array[i,0])

#converting to numpy array so that we can provide the data to LSTM model
x_train, y_train = np.array(x_train), np.array(y_train)

```

```python
#ML Model
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential

#creating the sequential model
model = Sequential()

#adding layers 

#first layer
#x.train.shape[1] indicates the 100 columns (which are shown as features) 
# and (x_train.shape[1], 1) is considering only one column i.e., closing col
model.add(LSTM(units = 50, activation = 'relu', return_sequences = True, 
               input_shape = (x_train.shape[1], 1)))
model.add(Dropout(0.2))

#2nd layer
model.add(LSTM(units = 60, activation = 'relu', return_sequences = True, ))
model.add(Dropout(0.3))

#3rd layer
model.add(LSTM(units = 80, activation = 'relu', return_sequences = True, ))
model.add(Dropout(0.4))

#4th layer
model.add(LSTM(units = 120, activation = 'relu', return_sequences = True, ))
model.add(Dropout(0.5))

#dense layer: we are predicting only 1 value
model.add(Dense(units = 1))

model.summary()

```

```python
model.compile(optimizer='adam', loss= 'mean_squared_error')
model.fit(x_train, y_train, epochs = 50)
```

```python
model.save('keras_model.h5')
```
